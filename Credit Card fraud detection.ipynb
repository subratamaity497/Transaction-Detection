# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

# Load the data to pandas data frame

```python
card_data=pd.read_csv('creditcard.csv')
```

# VIsualize the data

```python
card_data.head()
```

```python
card_data.info()
```

# How many value is null

```python
card_data.isnull().sum()
```

```python
card_data['Class'].value_counts()
```

# Spliting the data into fraud and legit

```python
legit=card_data[card_data.Class==0]
fraud=card_data[card_data.Class==1]
```

```python
legit.info()
```

```python
legit.head()
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

# As fraud transation data is very less we will perform under sampling on legit data

```python
legit_sample=legit.sample(n=500)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

To fit the data into machine learning model we will split into FEATUREs and TARGET

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

# Model Training

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

# Model Evalution

```python
X_train_prediction=model.predict(X_train)
train_accuracy=accuracy_score(X_train_prediction,Y_train)
print(train_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_accuracy=accuracy_score(X_test_prediction,Y_test)
print(test_accuracy)
```
